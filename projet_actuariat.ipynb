```python
#d√©finition des fonctions :

def lx(x,table):
    return table[x]

def qx(x,table):
    return (lx(x,table)-lx(x+1,table))/lx(x,table)

def dx(x,table):
    return lx(x,table) - lx(x+1,table)

def npx(x,n,table):
    return lx(x+n,table)-lx(x,table)

def vn(i,n):
    return 1/(1+i)^n

def nEx(x,i,n,table):
    return npx(x,n,table)*vn(i,n)

def h11qx(x,h,table):
    return (lx(x+h-1,table)-lx(x+h,table))/lx(x,table)

def h11Ax(x,h,i,table):
    return h11qx(x,h,table)*vn(i,h)

def Ax(x,h,i,table):
    Ax = 0
    for k in range(1,w-x+1):
        Ax += h11qx(x,k,i,table)
    return Ax
    
def nAx(x,n,i,table):
    nAx = 0
    for i in range(1,n+1):
        nAx += h11Ax(x,n,i,table)
    return nAx

def U_PE(x,i,n,benefit,table):
    return nEx(x,i,n,table)*benefit

def U_WL(x,i,n,benefit,table):
    return Ax(x,n,i,table)*benefit

def U_TA(x,i,n,benefit,table):
    return nAx(x,i,n,table)*benefit

def U_Endow(x,i,n,benefit,table):
    return (nAx(x,i,n,table)+nEx(x,i,n,table))*benefit

def axm(x,i,m,table):
    axm = 0
    for k in range(0,m):
        axm += nEx(x,i,k,table)
    return axm

def Annual_Premium_PE(x,i,n,benefit,m,table):
    return U_PE(x,i,n,benefit,table)/axm(x,i,m,table)

def Annual_Premium_WL(x,i,n,benefit,m,table):
    return U_WL(x,i,n,benefit,table)/axm(x,i,m,table)

def Annual_Premium_TA(x,i,n,benefit,m,table):
    return U_TA(x,i,n,benefit,table)/axm(x,i,m,table)

def Annual_Premium_Endow(x,i,n,benefit,m,table):
    return U_Endow(x,i,n,benefit,table)/axm(x,i,m,table)

```
